```python
import os
import cv2
import cvzone
from cvzone.ClassificationModule import Classifier

cap = cv2.VideoCapture(0)

classifier = Classifier(
    r"C:\MACHINE LEARNING PROJECTS\object Dedection\keras_model.h5",
    r"C:\MACHINE LEARNING PROJECTS\object Dedection\labels.txt"
)

imgArrow = cv2.imread(
    r"C:\MACHINE LEARNING PROJECTS\object Dedection\Resources (2)\Resources\Resources\arrow.png",
    cv2.IMREAD_UNCHANGED
)

classIDBin = 0

# Waste images load
imgWasteList = []
pathFolderWaste = r"C:\MACHINE LEARNING PROJECTS\object Dedection\Resources (2)\Resources\Resources\Waste"
for path in os.listdir(pathFolderWaste):
    imgWasteList.append(cv2.imread(os.path.join(pathFolderWaste, path), cv2.IMREAD_UNCHANGED))

# Bin images load
imgBinsList = []
pathFolderBins = r"C:\MACHINE LEARNING PROJECTS\object Dedection\Resources (2)\Resources\Resources\Bins"
for path in os.listdir(pathFolderBins):
    imgBinsList.append(cv2.imread(os.path.join(pathFolderBins, path), cv2.IMREAD_UNCHANGED))

classDic = {
    0: None,
    1: 0, 2: 0,
    3: 3, 4: 3,
    5: 1, 6: 1,
    7: 2, 8: 2
}

while True:
    success, img = cap.read()
    if not success:
        break

    imgResize = cv2.resize(img, (454, 340))
    imgBackground = cv2.imread(
        r"C:\MACHINE LEARNING PROJECTS\object Dedection\Resources (2)\Resources\Resources\background.png"
    )

    prediction = classifier.getPrediction(img)
    classID = prediction[1]
    print("ClassID:", classID)

    if classID != 0:
        imgBackground = cvzone.overlayPNG(imgBackground, imgWasteList[classID - 1], (909, 127))
        imgBackground = cvzone.overlayPNG(imgBackground, imgArrow, (978, 320))
        classIDBin = classDic[classID]

    if classIDBin is not None:
        imgBackground = cvzone.overlayPNG(imgBackground, imgBinsList[classIDBin], (895, 374))

    imgBackground[148:148 + 340, 159:159 + 454] = imgResize

    cv2.imshow("Image", img)
    cv2.imshow("Output", imgBackground)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
